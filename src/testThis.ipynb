```python
import matplotlib.pyplot as plt

import DiffusionFreeGuidence.DiffusionCondition as dc

start = 0.1
end = 0.9
steps = 1000
cosine_schedule = dc.cosine_linspace(start, end, steps)

plt.plot(cosine_schedule.numpy())
plt.xlabel('Steps')
plt.ylabel('Value')
plt.title('Cosine Schedule')
plt.show()
```
